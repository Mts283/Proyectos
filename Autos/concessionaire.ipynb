```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df_import = pd.read_csv("db.csv", sep=";")

df_concessionaire = pd.DataFrame(df_import)
```

```python
df_concessionaire
```

Cleaned DataFrame and zeros were added for "Kilometraje"

```python
df_concessionaire.isna().sum()
```

```python
df_clean = df_concessionaire.fillna(0)

df_clean
```

```python
df_clean.isna().sum()
```

Cleaned and split the column "Accesorios"

```python
df_clean["Accesorios"] = df_clean["Accesorios"].str.replace(r"\[|\]|'", "", regex=True)

df_clean
```

Split the column "Accesorios"

```python
accessories_dummies = df_clean["Accesorios"].str.get_dummies(sep=", ")

accessories_dummies
```

Combine both variables

```python
df_final = pd.concat([df_clean, accessories_dummies], axis=1)

df_final = df_final.drop(columns=["Accesorios"])

df_final.head()
```

```python
df_final.info()
```

Split the column "Motor"

```python
df_final = pd.get_dummies(df_final, columns=["Motor"])

df_final
```

Delete the column "Nombre"

```python
df_final = df_final.drop(columns="Nombre")

df_final
```

Change bool to int the column "Cero_km"

```python
df_final["Cero_km"] = df_final["Cero_km"].astype(int)

df_final
```

Create X e y

```python
X = df_final.drop(columns=["Valor"])

X
```

```python
y = df_final["Valor"]

y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

Training the IA

```python
train_forest = RandomForestRegressor(random_state=42)

train_forest.fit(X_train, y_train)
```

```python
forest_predict = train_forest.predict(X_test)

forest_predict
```

```python
error_df = mean_absolute_error(y_test, forest_predict)

error_df
```

Option A: Linear Regression

```python
linear_train = LinearRegression()

linear_train.fit(X_train, y_train)
```

```python
linear_predict = linear_train.predict(X_test)

linear_predict
```

```python
linear_error = mean_absolute_error(y_test, linear_predict)

linear_error
```

Option B - Gradient Boosting

```python
boosting_model = GradientBoostingRegressor(n_estimators=100, random_state=42)

boosting_model.fit(X_train, y_train)
```

```python
boosting_predict = boosting_model.predict(X_test)

boosting_predict
```

```python
boosting_error = mean_absolute_error(y_test, boosting_predict)

boosting_error
```
